```python
ori = np.array(decode.read_gtif(datadir + raw_f[index]))
im1 = plt.imshow(ori, cmap=cm.terrain)
plt.clim(0,500)
plt.colorbar(im1)
```

```python
err = abs(dcd - ori)
im = plt.imshow(err, cmap=cm.terrain)
plt.clim(0,500)
plt.colorbar(im)
```

```python

import matplotlib.ticker as mtick
fig = plt.figure(1, (7, 4))
ax = fig.add_subplot(1, 1, 1)
yticks = mtick.PercentFormatter(len(err.flatten()))
ax.yaxis.set_major_formatter(yticks)
ax.set_xlabel("Compressed differentials (m)")
ax.hist(err.flatten(), bins=range(0, 50))
```

```python

import csv

with open('data_out.csv', 'w', newline='') as file:
    writer = csv.writer(file)

    for index in range(len(files)):
        dcd = np.array(decode.decompress(inpath + files[index], 2048))
        ori = np.array(decode.read_gtif(datadir + raw_f[index]))
        err = abs(dcd - ori)

        diffs = err.flatten()
        mdf_aberations = np.delete(diffs, np.where(diffs > 50))
        compsize = os.stat(inpath + gpgc_f[index]).st_size
        orisize = os.stat(datadir + raw_f[index]).st_size

        print("FNAME\t", raw_f[index])
        print("ORISIZE\t", orisize)
        print("COMPSIZE", compsize)
        print("RATIO\t", orisize / compsize)
        print("STDDEV\t", mdf_aberations.std())
        print("MEAN\t", mdf_aberations.mean())

        writer.writerow([raw_f[index], gpgc_f[index], orisize, compsize, orisize / compsize, mdf_aberations.std(), mdf_aberations.mean()])
```
